# Поиск закономерностей успешности игр для прогнозиования будущих продаж

### Описание цели и задач

Для интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры, необходимо отработать принцип работы с данными, с помощью которого возможно составить прогноз продаж на будущий год на основе данных предыдущего года.

**Цель исследования** - выявить закономерности, определяющие успешность игры, чтобы сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Ход исследования**

Входные данные получены из открытых источников и представляют собой исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation) до 2016 года.

<p> В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».</p> 

Данные о пользователях получаем из файла `games.csv`.

Исследование пройдёт в шесть шагов:

1. Обзор данных
2. Предобработка данных
3. Проведение исследовательского анализа
4. Составление портрета пользователя каждого региона (NA, EU, JP)
5. Проверка гипотез
6. Общий вывод

**Описание данных**

Таблица `Игры` или `games.csv` состоит из следующих столбцов:

* `name` — название игры
* `platform` — платформа
* `year_of_release` — год выпуска
* `genre` — жанр игры
* `na_sales` — продажи в Северной Америке (миллионы проданных копий)
* `eu_sales` — продажи в Европе (миллионы проданных копий)
* `jp_sales` — продажи в Японии (миллионы проданных копий)
* `other_sales` — продажи в других странах (миллионы проданных копий)
* `critic_score` — оценка критиков (максимум 100)
* `user_score` — оценка пользователей (максимум 10)
* `rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Построим столбчатую диаграмму распределения игр по платформам по данным из столбца `'platform'`

```python
# Строим столбчатую диаграмму распределения количества игр по платформам по столбцу 'platform'
games.pivot_table(index='platform', values='name', aggfunc='count')\
    .sort_values(by='name')\
    .plot.barh(figsize=(10, 6),legend=False, color='#BBA5CC', grid=False)

# Подписываем оси и название фигуры
plt.xlabel('Количество игр для платформы, шт')
plt.ylabel('Название платформы для игр')
plt.title('Распределение игр по платформам')

# Выводим фигуру
plt.show()
```

Построим столбчатую диаграмму распределения игр по жанрам по данным из столбца `'genre'`

```python
# Строим столбчатую диаграмму распределения количества игр по платформам по столбцу 'genre'
games.pivot_table(index='genre', values='name', aggfunc='count')\
    .sort_values(by='name')\
    .plot.barh(figsize=(10, 6),legend=False, color='#DF6B7C', grid=False)

# Подписываем оси и название фигуры
plt.xlabel('Количество игр в жанре')
plt.ylabel('Название жанра')
plt.title('Распределение игр по жанрам')

# Выводим фигуру
plt.show()
```

Изучим, что означает рейтинг ESRB

Entertainment Software Rating Board (ESRB) — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»).

* **«E»** («Everyone») — **«Для всех»**: <p>Содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первоначально **«K-A»** ("Kids to Adults").</p>


* **«E10+»** («Everyone 10 and older») — **«Для всех от 10 лет и старше»**: <p>Проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови.</p>


* **«T»** («Teen») — **«Подросткам»**: <p>Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.</p>


* **«M»** («Mature») — **«Для взрослых»**: <p>Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику.</p>


* **«AO»** («Adults Only 18+») — **«Только для взрослых»**: <p>Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое.</p>


* **«RP»** («Rating Pending») — **«Рейтинг ожидается»**: <p>Продукт был отправлен в ESRB и ожидает присвоения рейтинга. </p>

Таким образом, рейтинг «K-A» можно заменить как «E»

2. Создадим датафрейм `df_platform`, в который запишем сводную таблицу, в которую внесем минимальный (`'min_year_of_release'`), максимальный (`'max_year_of_release'`) и медианный (`'median_year_of_release'`) года игр для заданной платформы и количество данных (`'count'`), по которым они расчитаны.
3. Создадим датафрейм `help_games`, в который запишем строки датафрейма с расчетом медианного года для каждой игры
4. Соединим три датафрейма по типу платформ и названию игры в датафрейм `lost_games`
5. Создадим функцию `filling_years`, которая заполнит пропуски в получившимся датафрейме `lost_games` в столбце с годом релиза. Она работает так: если медианный год по другим платформам для игры с указанным названием известен и он лежит в диапазоне минимального и максимального значения для соответствующего типа платформы, то присваиваем значению года релиза медианный год по имени игры, если нет, то присваиваем медианный год релиза по всем играм для соответствующей платформы.
6. Формируем список из новых значений годов релиза, которым заполняем пропуски в исходном датасете.

```python
# Создаём датафрейм lost_games, в который записываем строки датафрейма с пропусками в дате релиза
lost_games = games.loc[games['year_of_release'].isna()]

# Выводим первые пять строк созданного датафрейма
lost_games.head()
```

```python
# Создаём датафрейм df_platform, в который записываем сводную таблицу, 
# в которую вносим минимальный (`'min_year_of_release'`), максимальный (`'max_year_of_release'`) и 
# медианный (`'median_year_of_release'`) года игр для заданной платформы и количество данных (`'count'`), 
# по которому они расчитаны.
df_platform = games.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max', 'median', 'count'])

# Меняем названия столбцов
df_platform.columns = ['min_year_of_release', 'max_year_of_release', 'median_year_of_release', 'count']

# Обновляем индексы
df_platform = df_platform.reset_index()

# Выводим первые пять строк созданного датафрейма
df_platform.head()
```

```python
# Создаём датафрейм help_games, в который запиcываем строки датафрейма с расчетом медианного года для каждой игры
help_games = games.loc[~games['year_of_release'].isna()].groupby('name')['year_of_release'].median().reset_index()

# Меняем названия столбцов
help_games.columns = ['name', 'median_year_by_name']

# Выводим первые пять строк созданного датафрейма
help_games.head()
```

```python
# Соединяем все три датафрейма
lost_games = lost_games.merge(df_platform, on='platform', how='left').merge(help_games, on='name', how='left')

# Выводим первые пять строк созданного датафрейма
lost_games.head()
```

```python
# Выведем основную информацию о датафрейме lost_games для проверки соединения таблиц
lost_games.info()
```

```python
# Функция, которая возвращает год релиза для заполнения пропусков
# На вход передпется строка из датафрейма
def filling_years(row):
    # Сравниваем для игры с известным годом релиза на другой платформе медианное значение ее года релиза с
    # минимальным и максимальным годом для указанного типа платформы
    if row['min_year_of_release'] <= row['median_year_by_name'] <= row['max_year_of_release']:
        return  row['median_year_by_name']  # возвращаем медианное значение по названию игры
    else:
        return row['median_year_of_release'] # возвращаем медианное значение по типу платформы
```

```python
# Применяем функцию filling_years построчно к lost_games и результат помещаем в столбец 'year_of_release'
lost_games['year_of_release'] = lost_games.apply(filling_years, axis=1)
```

```python
# Заполняем пропуски в games списком из столбца 'year_of_release' датафрейма lost_games 
games.loc[(games['year_of_release'].isna()), 'year_of_release'] = list(lost_games['year_of_release'])
```

##### Распределение суммарных мировых продаж по платформам

Создадим датафрейм`relevant_data` по срезу данных из `games` за актуальный период.

```python
# Создаём датафрейм sony из общего games по списку платформ из sony_platform,
relevant_data = games.query('year_of_release >= 2013')
```

```python
# Выведем основную информацию о датафрейме relevant_data
relevant_data.info()
```

Для построения круговой диаграммы с процентами суммарных мировых продаж по типу платформы, соберем данные в `top_platform `, заменим три платформы с наименьшими продажами на 'Other'

```python
# Создадим переменную top_platform, в которую поместим столбцы с типом платформы 'platform' 
# и с общими мировыми продажами 'total_sales'
top_platform = relevant_data[['platform', 'total_sales']]

# Отберем три платформы с самыми низкими суммарными продажами
change = relevant_data.groupby('platform')['total_sales'].sum().sort_values(ascending=False)\
    .reset_index()['platform'].tail(3).unique()

# Заменим три платформы с самыми низкими суммарными продажами в top_platform на 'Other'
for name_platform in change:
    top_platform = top_platform.where(top_platform != name_platform, 'Other')

# Посчитаем процент суммарных мировых продаж по типу платформы
top_platform = top_platform.groupby('platform')['total_sales'].sum()\
    .apply(lambda x: x / relevant_data['total_sales'].sum() * 100).sort_values(ascending=False)
```

Построим круговую диаграмму для доли суммарных мировых продаж по типу платформы

По диаграмме можно сделать вывод, что больше всего продаж у игр для PS4 - 29%, на втором месте PS3 - 17%, на третьем XOne - 15%, четвертое место делят 3DS и X360 - у них по 13%.

Построим графики распределения суммарных продаж по годам за актуальный период в зависимости от типа платформы.

```python
# Создаем фигуру для размещения графиков
plt.figure(figsize=(11, 7))

# Строим графики по платформам
for type_platform in top_platform.index[-2::-1]:
    plt.plot('year_of_release', 'total_sales', data=relevant_data.query('platform == @type_platform').groupby('year_of_release')['total_sales'].sum().reset_index(), marker='', linewidth=2, label=type_platform)

# Настраиваем подписи значений на оси ox
plt.xticks(range(2013, 2017, 1))

# Подписываем оси и названия графиков
plt.xlabel('Количество продаж, млн шт')
plt.ylabel('Название платформы для игр')
plt.title('Распределение суммарных продаж по платформам')

# Отображение легенды
plt.legend() 

# Выводим созданную фигуру
plt.show()
```

#### Исследование продаж игр в зависимости от их жанра

Оценим распределение количества игр по жанрам за актуальный год. Для этого построим круговую диаграмму.

```python
# Создаем фигуру для размещения графиков
plt.figure(figsize=(11, 6))

# Круговая диаграмма по доли количества игр разных жанров в процентах
plt.pie(relevant_data.pivot_table(index='genre', values='name', aggfunc='count')\
    .sort_values(by='name')['name'], labels=relevant_data.pivot_table(index='genre', values='name', aggfunc='count')\
    .sort_values(by='name').index, autopct='%1.f%%')

plt.title('Доля количества выпущенных игр\nдля разных жанров за актуальный период')

# Выводим созданную фигуру
plt.show()
```

Среди производителей игр самыми популярными жанрами являются: Action, Role-Playing и Adventure.
Но какие жанры самые популярные по продажам?

Для построения круговой диаграммы с процентами суммарных мировых продаж по жанру игры, соберем данные в `top_genre`, заменим пять жанров с наименьшими продажами на *'Other'*

```python
# Создадим переменную top_genre, в которую поместим столбцы с жанром
# и с общими мировыми продажами 'total_sales'
top_genre = relevant_data[['genre', 'total_sales']]

# Отберем пять платформ с самыми низкими суммарными продажами
change = relevant_data.groupby('genre')['total_sales'].sum().sort_values(ascending=False)\
    .reset_index()['genre'].tail(5).unique()

# Заменим пять платформ с самыми низкими суммарными продажами в top_genre на 'Other'
for name_genre in change:
    top_genre = top_genre.where(top_genre != name_genre, 'Other')

# Посчитаем процент суммарных мировых продаж по типу платформы
top_genre = top_genre.groupby('genre')['total_sales'].sum()\
    .apply(lambda x: x / relevant_data['total_sales'].sum() * 100).sort_values(ascending=False)
```

Построим круговую диаграмму для доли суммарных мировых продаж по жанру

```python
# Создаем фигуру для размещения графиков
plt.figure(figsize=(11, 6))

# Круговая диаграмма по доли суммарных мировых продаж по типу платформы top_platform в процентах
plt.pie(top_genre, labels=top_genre.index, autopct='%1.f%%')

plt.title('Доля суммарных мировых продаж игр\n различных жанров за актуальный период')

# Выводим созданную фигуру
plt.show()
```

---
* По диаграмме можно сделать вывод, что больше всего продаж у игр жанра Action - 34%, на втором месте Shooter - 21%, на третьем Sports - 14%, четвертое место у Role-Playing - 13%.
* Таким образом, самый популярный жанр как по количеству игр, так и по общемировым продажам - Action.
* В топ-4 жанра также входят Shooter, Sports и Role-Playing.
* Хотя жанр Adventure входит в топ-4 самых популярных жанров выпускаемых игр, он составляет незначительную часть общемировых продаж.

### Портрет пользователя каждого региона (NA, EU, JP)

Построим круговые диаграммы для анализа самых популярных платформ, жанров игр по продажам в Северной Америке, Европе и Японии. Также построим круговые диаграммы для оценки распределения продаж по рейтингу игр от ESBR.

```python
# Создаем фигуру для размещения графиков
plt.figure(figsize=(16, 16))

# Создаем счетчик
count = 0

# Создаем цикл для построения круговых диаграмм по регионам
for region in ['na_sales', 'eu_sales', 'jp_sales']:
    # Создаем словарь регионов
    region_name = {'na_sales':'Северной Америке', 'eu_sales':'Европе', 'jp_sales':'Японии'}
    # Увеличиваем счетчик
    count += 1 
    # 1. Строим круговые диаграммы для отображения топ-5 самых популярных платформ   
    # Создадим переменную top_platform, в которую поместим столбцы с типом платформы 'platform' и с продажами по региону
    top_platform = relevant_data[['platform', region]]
    
    # Отберем шесть платформ с самыми низкими суммарными продажами    
    change = relevant_data.groupby('platform')[region].sum().sort_values(ascending=False)\
        .reset_index()['platform'].tail(6).unique()
    
    # Заменим шесть платформ с самыми низкими суммарными продажами в top_platform на 'Other'    
    for name_platform in change:
        top_platform = top_platform.where(top_platform != name_platform, 'Other')
    
    # Посчитаем процент продаж по типу платформы    
    top_platform = top_platform.groupby('platform')[region].sum()\
        .apply(lambda x: x / relevant_data[region].sum() * 100).sort_values(ascending=False)

    # Задаем цвета для диаграммы
    colors = [ '#FF994F','#ffad72', '#ffc295', '#ffd6b9', '#ffebdc', '#ffffff']

    # Задаем размещение диаграммы на фигуре   
    plt.subplot(3, 3, count)

    # Строим круговую диаграмму по доли продаж в регионе для разных типов платформы top_platform в процентах
    plt.pie(top_platform, labels=top_platform.index, autopct='%1.f%%', colors=colors);
    
    # Подписываем названия диаграмм
    plt.title(f'Доля продаж игр в {region_name[region]}\nдля разных платформ за актуальный период');
    
    # 2. Строим круговые диаграммы для отображения топ-5 самых популярных жанров игр   
    # Создадим переменную top_genre, в которую поместим столбцы с жанром и с продажами в регионе
    top_genre = relevant_data[['genre', region]]
    
    # Отберем пять жанров с самыми низкими суммарными продажами
    change = relevant_data.groupby('genre')[region].sum().sort_values(ascending=False)\
        .reset_index()['genre'].tail(7).unique()
    
    # Заменим пять жанров с самыми низкими суммарными продажами в top_genre на 'Other'
    for name_genre in change:
        top_genre = top_genre.where(top_genre != name_genre, 'Other')
        
    # Посчитаем процент продаж в регионе по жанру
    top_genre = top_genre.groupby('genre')[region].sum()\
        .apply(lambda x: x / relevant_data[region].sum() * 100).sort_values(ascending=False)
    
    # Задаем цвета для диаграммы
    colors = ['#2ea7f2', '#58b9f5', '#82caf7', '#abdcfa', '#d5edfc', '#ffffff']
    
    # Задаем размещение диаграммы на фигуре
    plt.subplot(3, 3, count + 3)
    
    # Строим круговую диаграмму по доли продаж в регионе для разных жанров игр top_genre в процентах
    plt.pie(top_genre, labels=top_genre.index, autopct='%1.f%%', colors=colors);
    
    # Подписываем названия диаграмм
    plt.title(f'Доля продаж игр в {region_name[region]}\n различных жанров за актуальный период');
    
    # 3. Строим круговые диаграммы для отображения рейтинга ESBR по доле продаж в регионе
    # Создаем сводную таблицу ESBR_rating с долей продаж игр в регионе в зависимости от рейтинге ESBR
    ESBR_rating = relevant_data.groupby('rating')[region].sum()\
        .apply(lambda x: x / relevant_data[region].sum() * 100).sort_values(ascending=False)
    
    # Задаем цвета для диаграммы
    colors = ['#08a848', '#46be76', '#84d4a4', '#c1e9d1', '#ffffff']
    
    # Задаем размещение диаграммы на фигуре
    plt.subplot(3, 3, count + 6)
    
    # Строим круговую диаграмму по доли продаж в регионе для разных жанров игр top_genre в процентах
    plt.pie(ESBR_rating, labels=ESBR_rating.index, autopct='%1.f%%', colors=colors)
    
    # Подписываем названия диаграмм
    plt.title(f'Распределение продаж игр\nпо рейтингу ESBR в {region_name[region]}')
    
# Выводим созданные фигуры
plt.show()
```

### Проверка гипотез

#### Проверка гипотезы о равенстве среднего рейтинга для Xbox One и PC

Проверим, как выглядит распределение пользовательского рейтинга для платформ Xbox One и PC

```python
# Строим гистограмму распределения пользовательского рейтинга для платформы PC
ax = relevant_data.query('platform == "PC"')\
    .plot(kind='hist',y='user_score', bins=40, figsize=(10, 5), alpha=0.4, label='PC')

# Строим гистограмму распределения пользовательского рейтинга для платформы Xbox One
relevant_data.query('platform == "XOne"')\
    .plot(kind='hist',y='user_score', bins=40, alpha=0.4, label='Xbox One', ax=ax)

# Подписываем графики
plt.xlabel('Рейтинг пользователей')
plt.ylabel('Количество пользователей')
plt.title('Гистограмма распределения рейтинга пользователей для PC и Xbox One')

# Выводим созданную фигуру
plt.show()
```

```python
# Вычисляем среднее значение рейтинга пользователей для PC
avg = round(relevant_data.query('platform == "PC"')['user_score'].mean(), 2)
print(f'Среднее значение рейтинга пользователей для PC {avg}')

# Вычисляем среднее значение рейтинга пользователей для XOne
avg = round(relevant_data.query('platform == "XOne"')['user_score'].mean(), 2)
print(f'Среднее значение рейтинга пользователей для XOne {avg}')
```

Так как рассматриваем две независимые выборки, то используем `ttest_ind` для получения статистического критерия и p-value.

Проверим гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, то есть сформулируем нулевую и альтернативную гипотезы и проверим:

**Нулевая гипотеза:** средние значения пользовательских рейтингов платформ Xbox One и PC равны.

**Альтернативная гипотеза:** среднее значение рейтинга пользователей для PC меньше, чем среднее значение рейтинга пользователей для XOne.

Исходя из заданных нулевой и альтернативной гипотезы, применяем односторонний Т-тест (`alternative='less'`). Так как по гистограмме для распределения рейтинга пользователей для PC и Xbox One заметно, что они имеют разную дисперсию, то используем `equal_var=False`, параметр `nan_policy='omit'` позволяет исключить из выборки NaN. Уровень статистической значимости 5%

```python
# уровень статистической значимости
alpha = 0.05

# проводим статистический тест двух независимых выборок
results = st.ttest_ind(relevant_data.query('platform == "PC"')['user_score'],
                       relevant_data.query('platform == "XOne"')['user_score'],
                       equal_var=False, alternative='less', nan_policy='omit')

# выводим p-value и статистику
print(f'p-value: {results.pvalue: .2f}')
print(f't-критерий: {results.statistic: .2f}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Принимаем нулевую гипотезу')
```

*Нулевую гипотезу о том, что средние значения пользовательских рейтингов платформ Xbox One и PC равны, принимаем с уровнем значимости 5%*
